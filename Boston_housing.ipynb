## 3) Data Cleaning

- This function prints the sum of NA values in each column.

```r
sapply(df.boston, function(z) sum(is.na(z)))
```

- Remove unwanted dataset as a part of data cleaning.
- Removed columns are: chas, concetration of nitrogenmonoxide, black.

```r
df.updated <- subset(df.boston, select = -c(X,chas,nox, black))
head(df.updated)
```

- Rename the column names to better fit the data, as it makes more readable.

```r
colnames(df.updated)[1] <- "crime_rate"
colnames(df.updated)[2] <- "residential_land"
colnames(df.updated)[3] <- "industrial_acres"
colnames(df.updated)[4] <- "rooms"
colnames(df.updated)[6] <- "average_distance_to_employment"
colnames(df.updated)[7] <- "accessibility_to_highways"
colnames(df.updated)[11] <- "median_value_of_house"
colnames(df.updated)[10] <- "lower_status_population"
head(df.updated)
```

## 9) Multiple linear 

- This is the multi-linear regression of model 1 which predicts the price of the house per room against tax, non-retail industrial acres, number of rooms and age of the house.
<center> $ Y = -1.8X1 - 5.2X2 + 545.0X3 - 6.2X4 + 1339.1 $ </center>

```r
mulline1 <- lm(house_price_per_room ~ tax + industrial_acres + rooms + age, data =df.updated )
summary(mulline1)
```

```r
summary(mulline1)$coefficient
```

- This is the multi-linear regression model 2 which predicts the price of the house per room against lower status population, student-to-teacher ratio and crime rate.
<center> $ Y = -94.8X1 - 120.7X2 - 11.5X3 + 6996.5 $ </center>

```r
mulline2 <- lm(house_price_per_room ~ lower_status_population + ptratio + crime_rate, data =df.updated )
summary(mulline2)
```

```r
summary(mulline2)$coefficient
```

- This code display the output of the two multi-linear regression model.
- It is clearly seen that Model 2 out-performs Model 1 and also has higher **Adjusted r-sqaured of 0.536**


```r
cat("Adjusted R-squared value of Model 1:", summary(mulline1)$adj.r.squared,
    "\nAdjusted R-squared value of Model 2:", summary(mulline2)$adj.r.squared)
```
